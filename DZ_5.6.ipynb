```python
print("Начало игры!")
print("Вводите ходы в виде x y, где x - номер столбца, y - номер строки")
print('')
field = [['', '1', '2', '3'],
['1', '-', '-', '-'],
['2', '-', '-', '-'],
['3', '-', '-', '-']]

for line in field:
    col_0, col_1, col_2, col_3 = line
    print(f'{col_0:3} {col_1:3} {col_2:3} {col_3:3}')
print('')
    
hody = []

def process(move):
    if min(move) < 1 or max(move) > 3:
        print("Неверные координаты")
    elif move in hody:
        print("Поле занято")
    else:
        hody.append(move)
    
        if len(hody) % 2:
            field[move[1]][move[0]] = 'X'
        else:
            field[move[1]][move[0]] = 'O'

    for line in field:
        col_0, col_1, col_2, col_3 = line
        print(f'{col_0:3} {col_1:3} {col_2:3} {col_3:3}')

        
def victory_check(pole):
    for i in range (1, 4):
        if pole[i][1] == pole[i][2] == pole[i][3] != "-":
            print(f"{pole[i][1]} выиграли")
            return True
            break
    for j in range (1, 4):
        if pole[1][j] == pole[2][j] == pole[3][j] != "-":    
            print(f"{pole[1][j]} выиграли")
            return True
            break
    if pole[1][1] == pole[2][2] == pole[3][3] != "-" or pole[1][3] == pole[2][2] == pole[3][1] != "-":
        print(f"{pole[2][2]} выиграли")
        return True
        
def hod():
    move = list(map(int, input("Следующий ход").split()))
    return move

count = 0

while True:
    count += 1
    
    X = hod()
    print('')
    process(X)
       
    if victory_check(field):
        break
    
    if count == 9:
        print("Ничья!")
        break
```
